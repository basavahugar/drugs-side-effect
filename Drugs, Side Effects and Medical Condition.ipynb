```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = "drugs.csv"
data = pd.read_csv(file_path)
```

```python
# Display basic information and preview the data
print("Dataset Info:")
data.info()
```

```python
print("\nDataset Preview:")
print(data.head())
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("\nMissing Values:")
print(missing_values)
```

```python
# Top 10 drugs with the highest number of reviews
top_drugs = data[['drug_name', 'no_of_reviews']].dropna().sort_values(by='no_of_reviews', ascending=False).head(10)
print("\nTop 10 Drugs by Number of Reviews:")
print(top_drugs)
```

```python
# Most common medical conditions
top_conditions = data['medical_condition'].value_counts().head(10)
print("\nTop 10 Medical Conditions:")
print(top_conditions)
```

```python
# Plot the distribution of ratings
plt.figure(figsize=(8, 6))
sns.histplot(data['rating'], bins=20, kde=True, color='blue')
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# Bar plot for top 10 medical conditions
plt.figure(figsize=(10, 6))
top_conditions.plot(kind='bar', color='green')
plt.title("Top 10 Medical Conditions")
plt.xlabel("Medical Condition")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average rating per medical condition
avg_rating_per_condition = data.groupby('medical_condition')['rating'].mean().sort_values(ascending=False).head(10)
print("\nAverage Rating per Medical Condition:")
print(avg_rating_per_condition)
```

```python
# Plot the top 10 average ratings per medical condition
avg_rating_per_condition.plot(kind='bar', figsize=(10, 6), color='orange')
plt.title("Top 10 Average Ratings per Medical Condition")
plt.xlabel("Medical Condition")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```
